```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

First let's set the parametes:  
$\rho$=0.95  
$\sigma$=0.007

```python
rho=0.95
sigma=0.007
n=101
```

Now we'll find the upper and lower bound of the grid, by Tauchen's method.  
$$\theta_N=m\frac{\sigma}{\sqrt{1-\rho^2}}$$  
$$\theta_1=-m\frac{\sigma}{\sqrt{1-\rho^2}}$$
To be conservative I will use m = 3  


```python
m=3
upp=m*sigma/(1-rho**2)**0.5
low=-upp
```

Now let's generate equidistant points, as well as the "separators"

```python
xgrid = np.linspace(low, upp,n )


def grid(a,b,n):
    w=((np.sqrt(b-a))**2)/(n-1)
    x=np.empty(n-1)
    x[0]=a+w/2
    x[n-2]=b-w/2
    for i in range (n-3):
        x[i+1]=x[i]+w
    return x
            
y=grid(low,upp,n)
```

Now we can finally compute the cdfs of the limits we have created for our state spaces, then create the Markov chain of the process

```python
from scipy.stats import norm


trm = np.zeros((n, n-1))
for j in range(n-1):
    for i in range(n):     
        trm[i, j] = norm.cdf(y[j],loc=rho*xgrid[i],scale=sigma)
trmo=np.zeros((n,n))
for i in range(n): 
    trmo[i,0]=trm[i,0]
    trmo[i,n-1]=1-trm[i,n-2]
    for j in range(n-2):
        trmo[i,j+1]=trm[i,j+1]-trm[i,j]
```

 now simulating the process

```python
import random

random.seed(1658)
```

```python
import random
np.random.seed(1658)
nsteps = 1000
N=int((n-1)/2)
current_state =  N
simulated_states = np.zeros(nsteps)
simulated_states[0] = 0

# Transition matrix

# Simulation loop
for i in range(1, nsteps):
    current_state = np.random.choice(n, p=trmo[current_state])
    simulated_states[i] = xgrid[current_state]
```

```python
plt.plot(simulated_states)
plt.show
```

```python
np.random.seed(1658)
# Generate shocks
shocks = np.random.normal(scale=sigma, size=nsteps)

# Use the shocks to generate the states of the AR(1) process
x = np.empty(nsteps + 1)
x[0] = 0
alpha = 0.95
for t in range(nsteps):
    x[t+1] = x[t]*alpha + shocks[t]
current_state=np.empty(nsteps+1)
current_state[0]=0
for i in range(nsteps):
    current_state[i+1] = xgrid[np.abs(xgrid - (current_state[i]*alpha + shocks[i])).argmin()]
```

```python
maxi = (sigma**2 / (1 - rho**2))**(1/2) * np.sqrt(n - 1)
mini=-maxi
xgrid2 = np.linspace(mini, maxi, n)
```

```python
def rouwenhorst(n, rho):
    p = (1 + rho) / 2

    if n == 2:
        theta = np.array([[p, 1 - p], [1 - p, p]])

    else :
        p1 = np.zeros((n, n))
        p2 = np.zeros((n, n))
        p3 = np.zeros((n, n))
        p4 = np.zeros((n, n))

        new_mat = rouwenhorst(n - 1, rho)

        p1[:n - 1, :n - 1] = p * new_mat
        p2[:n - 1, 1:] = (1 - p) * new_mat
        p3[1:, :-1] = (1 - p) * new_mat
        p4[1:, 1:] = p * new_mat

        theta = p1 + p2 + p3 + p4
        for i in range (n):
            theta[i: ,:] = (theta[i:,:] / sum(theta[i]))

    return theta

N=rouwenhorst(n, 0.95,)
```

```python
np.random.seed(1658)
z=int((n-1)/2)
current_state =  z
estados = np.zeros(nsteps)
estados[0] = 0

# Transition matrix

# Simulation loop
for i in range(1, nsteps):
    current_state = np.random.choice(n, p=N[current_state])
    estados[i] = xgrid2[current_state]
plt.plot(estados)
plt.show()
```
